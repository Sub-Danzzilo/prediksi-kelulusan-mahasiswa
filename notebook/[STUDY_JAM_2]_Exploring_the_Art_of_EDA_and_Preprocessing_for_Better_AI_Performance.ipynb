## Data Cleaning

###### Handling missing value

```python
df = df.dropna(subset=['IPS_8']).copy()
```

###### Handling data inconsistencies

```python
df['JENIS_KELAMIN'] = df['JENIS_KELAMIN'].replace({
    'MALE': 'LAKI - LAKI',
    'WANITA': 'PEREMPUAN'
})
```

```python
df['JENIS_KELAMIN'].value_counts()
```

###### Handling data duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.shape
```

## Data Preprocessing

###### Encoding of data

```python
df['JENIS_KELAMIN'] = df['JENIS_KELAMIN'].map({
    'LAKI - LAKI': 1,
    'PEREMPUAN': 0
})
```

```python
df['STATUS_MAHASISWA'] = df['STATUS_MAHASISWA'].map({
    'BEKERJA': 1,
    'TIDAK BEKERJA': 0
})
```

```python
df['STATUS_NIKAH'] = df['STATUS_NIKAH'].map({
    'MENIKAH': 1,
    'BELUM MENIKAH': 0
})
```

```python
df['STATUS_KELULUSAN'] = df['STATUS_KELULUSAN'].map({
    'TEPAT': 1,
    'TERLAMBAT': 0
})
```

```python
df.head()
```

```python
df.info()
```

###### Feature Selection

```python
df = df.drop(['NAMA','IPS_5', 'IPS_6', 'IPS_7', 'IPS_8', 'IPK'], axis=1)
```
